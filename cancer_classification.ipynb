```python
X = df.drop('diagnosis',axis = 'columns')
y = df.diagnosis

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3 , shuffle = True , random_state = 44)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(penalty = 'l2' , solver = 'lbfgs' , C = 10 , tol = 0.2,max_iter = 15000)
LR.fit(X_train,y_train)
y_pred = LR.predict(X_test)
Score_LR = LR.score(X_test,y_test)
print(Score_LR)
```

```python
#K-Neighbour Classification
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors = 3, weights = 'distance' , algorithm = 'auto')
KNN.fit(X_train,y_train)
y_pred = KNN.predict(X_test)
Score_KNN = KNN.score(X_test,y_test)
print(Score_KNN)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
Decision = DecisionTreeClassifier(max_depth = 7 , min_samples_split = 3 )
Decision.fit(X_train,y_train)
y_pred = Decision.predict(X_test)
Score_Decision = Decision.score(X_test,y_test)
print(Score_Decision)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
Random = RandomForestClassifier(n_estimators = 7 , max_depth = 4 , min_samples_split = 2)
Random.fit(X_train,y_train)
y_pred = Random.predict(X_test)
Score_Random = Random.score(X_test,y_test)
print(Score_Random)
```

```python
#XGBoost
from sklearn.ensemble import GradientBoostingClassifier
GBC = GradientBoostingClassifier(n_estimators = 23 , learning_rate = 0.1)
GBC.fit(X_train,y_train)
y_pred = GBC.predict(X_test)
Score_GBC = GBC.score(X_test,y_test)
print(Score_GBC)
```

```python
#Support Vector Classifier (SVC)
from sklearn.svm import SVC
svc = SVC(kernel = 'linear' , tol= 0.1 , C = 0.4,max_iter = 1000)
svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)
Score_svc = svc.score(X_test,y_test)
print(Score_svc)
```

```python
#Multi-Layer Perception (MLP)
from sklearn.neural_network import MLPClassifier
neural = MLPClassifier(activation = 'logistic' , hidden_layer_sizes = 5 , solver = 'lbfgs'  , alpha= 0.1,random_state = 44,max_iter = 1000)
neural.fit(X_train,y_train)
y_pred = neural.predict(X_test)
Score_neural = neural.score(X_test,y_test)
print(Score_neural)
```
